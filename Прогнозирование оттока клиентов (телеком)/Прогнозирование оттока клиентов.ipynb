# Описание проекта
Оператор связи ***«Ниединогоразрыва.ком»*** хочет научиться прогнозировать отток клиентов. Если выяснится, что пользователь планирует уйти, ему будут предложены промокоды и специальные условия. Команда оператора собрала персональные данные о некоторых клиентах, информацию об их тарифах и договорах.
 
# Цель проекта  
Разработка модели машинного обучения - подбор оптимального алгоритма и гиперпараметров для решения задачи классификации по прогнозированию оттока клиентов (попадания в чёрн), исходя из имеющихся признаков клиентов. Оптимальной считается модель, показавшая значение метрики качества ROC-AUC на уровне не меньше 0.85. Если таких моделей более одной - выбираем ту, которая обучается быстрее.

# Описание услуг
Оператор предоставляет два основных типа услуг: 
1. Стационарную телефонную связь. Возможно подключение телефонного аппарата к нескольким линиям одновременно.  
2. Интернет. Подключение может быть двух типов: через телефонную линию (<i><b>DSL</b></i>, от англ. digital subscriber line, «цифровая абонентская линия») или оптоволоконный кабель (<i><b>Fiber optic</b></i>).  

Также доступны такие услуги:
- Интернет-безопасность: антивирус (<i><b>DeviceProtection</b></i>) и блокировка небезопасных сайтов (<i><b>OnlineSecurity</b></i>);  
- Выделенная линия технической поддержки (<i><b>TechSupport</b></i>);  
- Облачное хранилище файлов для резервного копирования данных (<i><b>OnlineBackup</b></i>);  
- Стриминговое телевидение (<i><b>StreamingTV</b></i>) и каталог фильмов (<i><b>StreamingMovies</b></i>).  

За услуги клиенты могут платить каждый месяц или заключить договор на 1–2 года. Доступны различные способы расчёта и возможность получения электронного чека.

#### 2.2. Работа с признаками

Попробуем оценить изменение тарифа клиентами. Отнимем от месячных расходов умноженных на количество месяцев пользования услугами, суммарные расходы клиента за всё время.  

Если полученная сумма больше нуля - значит тариф был изменен в сторону увеличения оплаты (куплены дополнительные услуги или расширены пакеты).  
Если полученная сумма меньше нуля - значит тариф был изменен в сторону более дешевого, либо сокращен объем приобретаемых услуг.  

Так же не стоит упускать из виду, что цены могли меняться независимо от предпочтений клиента - выгодные акции на новые тарифы с аналогичным набором услуг, или просто небольшое подорожание из-за инфляции. Ну и наш расчёт определенно не будет избавлен от погрешностей. Всё это будет одна большая категория тех, кто не менял условия обслуживания.

```python
df['DevTotal'] = (df['MonthlyCharges'] * ((df['Duration'] / 30).round())).round(2) - df['TotalCharges']
```

```python
df['DevTotal'].describe(percentiles=[i for i in np.arange(0.1, 1, 0.1)])
```

Видно, что гораздо больше клиентов сменили тариф на более дешевый, чем на более дорогой.  

Изучим распределение получше.

```python
plt.figure(figsize=(10, 6))
plt.hist(df['DevTotal'], bins=100)
plt.xticks([i for i in range(-750, 155, 50)])
plt.xlim(-750, 150)
plt.title('Распределение изменения платы за услуги связи', fontsize=15)
plt.grid(True)
plt.show()
```

Кажется расчет оказался точнее, чем казалось изначально. Посмотрим какая доля "нулевых" отклонений.

```python
print('Доля нулевых отклонений от общего числа клиентов = {:.1%}'.format(df.loc[df['DevTotal'] == 0].shape[0] / df.shape[0]))
```

Почти 30%. Отметим так же как не изменявших тариф тех, у кого отклонение оказалось менее 10. Дань ошибке.

```python
df['PayUp'] = df['DevTotal'].apply(lambda x: 'Yes' if x > 10 else 'No')
df['PayDown'] = df['DevTotal'].apply(lambda x: 'Yes' if x < -10 else 'No')
```

### 3. Обучение моделей

#### 3.1. Подготовка данных

```python
RANDOM_STATE = 250923
```

Корректно. Теперь закодируем признаки - обучим OHE и стандартизатор.

```python
scaler = StandardScaler()

ohe_enc = OneHotEncoder(handle_unknown='ignore', drop='first')
```

```python
features_train_dummies = pd.DataFrame(
    ohe_enc.fit_transform(features_train.drop(columns=interval_data_columns)).toarray()
    ).join(pd.DataFrame(scaler.fit_transform(features_train[interval_data_columns].values)
                        ).rename(columns={0: 30, 1: 31}))
```

Сделано. Теперь приступим к поиску и обучению оптимальной модели.

#### 3.2. Поиск и обучение оптимальной модели

Для удобства напишем функцию, которая будет получать на вход необходимые переменные, обучать модели и перебирать гиперпараметры. По итогу будет выводить результат.

```python
def model_result(grid_model, features, target):
    grid_model.fit(features, target)
    print('Лучший результат ROC-AUC = {:.2f}'.format(grid_model.best_score_)) 
    print('Время обучения модели = {:.5f} сек.'.format(grid_model.cv_results_['mean_fit_time'].mean()))
    print('Лучшие параметры модели =', grid_model.best_params_)
    return grid_model.best_estimator_
```

Начнем с логистической регрессии.

```python
logistic_reg = LogisticRegression(random_state=RANDOM_STATE)

grid_lr = GridSearchCV(estimator=logistic_reg, n_jobs=-1, 
                       param_grid={'solver': ['sag', 'saga', 'liblinear']}, 
                       scoring='roc_auc')
```

```python
logistic_reg = model_result(grid_lr, features_train_dummies, target_train)
```

Результат не отягивает до необходимого.  
На очереди решающее дерево.

```python
dec_tree = DecisionTreeClassifier(random_state=RANDOM_STATE)

grid_dt = GridSearchCV(estimator=dec_tree, n_jobs=-1, 
                       param_grid={'max_depth': [i for i in range(2, 12)], 
                                   'min_samples_split': [2, 3, 4]}, 
                       scoring='roc_auc')
```

```python
dec_tree = model_result(grid_dt, features_train_dummies, target_train)
```

Лучше но по-прежнему недостаточно.  
Посмотрим что скажет градиентный бустинг.

```python
catboost_classes = CatBoostClassifier(random_state=RANDOM_STATE, verbose=0)

grid_cc = GridSearchCV(estimator=catboost_classes, n_jobs=-1, 
                       param_grid={'n_estimators': [i for i in range(20, 120, 10)], 
                                      'max_depth': [i for i in range(2, 12)], 
                                  'learning_rate': [i for i in np.arange(0.1, 1.0, 0.1)]}, 
                       scoring='roc_auc')
```

```python
%%time
catboost_classes = model_result(grid_cc, features_train_dummies, target_train)
```

Отлично. Градиентный бустинг показывает самый высокий результат.

#### 3.3. Проверка оптимальной модели на тестовых данных

Теперь применим модель к тестовым данным. Для начала стандартизируем и закодируем признаки.

Результат по-прежнему высок. Говорить о переобучении, думаю, не стоит.
